```python
import os
import pandas as pd
```

```python
path = os.getcwd()
```

```python
file_path = path + '/' + 'rfe_data.csv'
data = pd.read_csv(file_path)

data
```

```python
features = data.drop(columns=['SUB_ID', 'DX_GROUP', 'FILE_ID', 'EYE_STATUS_AT_SCAN',
       'HANDEDNESS_L', 'HANDEDNESS_R'])
labels = data['DX_GROUP']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
features = scaler.fit_transform(features)
```

```python
from model import train_and_evaluate
from plotting import plot_training_history, plot_confusion_matrix, plot_roc_curves

hyperparameters = {
    'units': 96,
    'l2_reg': 0.002242865877928494,
    'dropout_rate': 0.3,
    'learning_rate': 0.003956370567137552,
    'epochs': 50,
    'batch_size': 32
}

model_filename = 'best_model_Multimodal_RFE.h5'

fold_accuracies, fold_conf_matrices, fold_roc_curves, fold_auc_scores, history = train_and_evaluate(
    features, labels, hyperparameters, model_filename
)

plot_training_history(history)
plot_confusion_matrix(fold_conf_matrices, ['Class 1', 'Class 2'])
plot_roc_curves(fold_roc_curves, fold_auc_scores)
```
